```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
z=np.array([50,40,30,20,10])

# plt.plot(x,y, marker="o", mfc="green", mec="red", ms=20)
plt.xlabel("i am x")
plt.ylabel("i am y")
plt.title("hello world")
plt.grid(axis="y")
plt.subplot(2,1,1)
plt.plot(x,y,"sy",ls="--", lw=3, ms=10)

plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
x = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0.1, 0, 0.3]
myshadow = [0.2, 0.1, 0, 0.3]

plt.pie(x,labels=mylabels,startangle=90,explode=myexplode,shadow=myshadow)
plt.legend(title="fruits")
plt.show()
```
